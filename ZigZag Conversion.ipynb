```python
class Solution(object):
      
            
    def convert(self, s, numRows):
        """
        :type s: str
        :type numRows: int
        :rtype: str
        """
        sLen = len(s)        
        nBet = max(0,numRows-2)        
        
        if sLen == "":
            return ""
        elif numRows>sLen:
            return s
        
        # initialize initial string        
        strList = list(' '*sLen)
        
        # Fill in the initial verticals:        
        ndx = -1
        strList[ndx] = s[0]
        sval = 0            
        # First line contains only full vertical characters
        while sval<sLen:        
            ndx += 1
            strList[ndx] = s[sval]
            sval += numRows + nBet

        # Save off the total number of veritical characters
        nVerts = ndx+1
                
        # Iterate through each row
        for iRow in range(1,numRows-1):
            sC = iRow
            ndx += 1
            strList[ndx] = s[sC]
            
            # Check for characters immediately to the right of the veritical
            for iCol in range(1, nVerts+1):
                sR = sC + 2*(numRows - iRow -1)
                if  ndx<sLen and sR<sLen:
                    ndx += 1
                    strList[ndx] = s[sR]
                    
                sC += numRows + nBet
                if ndx<sLen and sC<sLen:
                    ndx+= 1
                    strList[ndx] = s[sC]

        # Iterate through last row with only full vertical characters
        sval = numRows-1
        for iRow in range(1, nVerts+1):
            if ndx<sLen-1 and sval<sLen:
                ndx += 1
                strList[ndx] = s[sval]
            sval += numRows + nBet

        
        newStr = ''.join(strList)
        return newStr
                
        
            
        
s = "PAYPALISHIRING"        
# s = "twckwuyvbihajbmhmodminftgpdcbquupwflqfiunpuwtigfwjtgzzcfofjpydjnzqysvgmiyifrrlwpwpyvqadefmvfshsrxsltbxbziiqbvosufqpwsucyjyfbhauesgzvfdwnloojejdkzugsrksakzbrzxwudxpjaoyocpxhycrxwzrpllpwlsnkqlevjwejkfxmuwvsyopxpjmbuexfwksoywkhsqqevqtpoohpd"
# s = "twckwuyvbihajbmhmodminftgpdcbquupwflqfiunpuwtihpd"
# s = ""
# s = "a"
# s  = "bacp"
numRows = 2
sObj = Solution()
sObj.convert(s,numRows)
```
