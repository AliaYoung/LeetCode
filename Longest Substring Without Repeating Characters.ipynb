```python
# Solution adapted from lightmark's brilliant c++ code
# Runs in O(n) time

class Solution(object):
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        maxLen = 0
        nStart = -1
        i = 0
        dict = {}
        while(not i == len(s)): # Iterate through all the possible lengths
        
            #print (i, nStart, dict)
            
            if s[i] in dict and dict[s[i]]> nStart:
                nStart = dict[s[i]]      # This is the brilliant piece!
            dict[s[i]] = i               # This is the brilliant piece!
            maxLen = max(maxLen, i-nStart)
            i += 1           
        return maxLen
        
# Testing the code        
str = "mycujjbgsgzldhszqciyjyqmpfnsfuqruunfgolyvmlavezlihjtiafxfupmohdkdqp"
sObj = Solution()
sObj.lengthOfLongestSubstring(str)
```
